# Sine Testing

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import numpy as np
import matplotlib.pyplot as plt
```

```python
length = 500

t = np.arange(500) / length

reference = np.cos(2*np.pi*t) - 1j * np.sin(2*np.pi*t)
```

```python
plt.plot(reference.real)
plt.plot(reference.imag)
```

## Sine value

```python
sine = np.sin(2*np.pi*t)
2 * np.abs(np.sum(sine * reference)) / len(reference)
```

## Zero

```python
2 * np.abs(np.sum(0 * reference)) / len(reference)
```

## Constant

```python
2 * np.abs(np.sum(-1 * reference)) / len(reference)
```

## Noise

```python
noise = np.random.normal(0, 0.3, size=500)
plt.plot(noise)
2 * np.abs(np.sum(noise * reference)) / len(reference)
```

# Sine with noise

```python
plt.plot(noise + sine)
2 * np.abs(np.sum((noise + sine) * reference)) / len(reference)
```
